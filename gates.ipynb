```python
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer
```

```python
# Creates a 1-qubit quantum circuit
qc = QuantumCircuit(1)

# Apply Gates
qc.x(0) # Apply X gate to qubit 0
qc.h(0) # Apply Hadamard gate to qubit 0
qc.z(0) # Apply Z gate to qubit 0
qc.s(0) # Apply S gate to qubit 0
qc.t(0) # Apply T gate to qubit 0
qc.rx(1.57, 0) # Apply RX gate with angle 0.5 to qubit 0
qc.measure_all() # Measure all qubits

qc.draw('mpl') # Draw the circuit using Matplotlib
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

# Create a simple quantum circuit
qc = QuantumCircuit(1, 1)
qc.h(0)
# qc.measure(0, 0)
qc.measure_all()

# Initialize the simulator backend
simulator = AerSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(qc, simulator)

# Run the circuit on the simulator
job = simulator.run(compiled_circuit, shots=1024)
result = job.result()

# Get and print the counts
counts = result.get_counts()
print(counts)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].bar(counts.keys(), counts.values(), width=0.5)
ax[0].set_title('Measurement Results')
ax[0].set_xlabel('Bitstring')
ax[0].set_ylabel('Counts')
ax[0].grid(True)


qc.draw('mpl', ax=ax[1])
plt.tight_layout()
plt.show()  # Display the plots
```
