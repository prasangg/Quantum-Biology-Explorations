```python
import numpy as np
from qutip import Bloch, basis

# Create a Bloch sphere object
b = Bloch()

# Define |0> and |1> basis states
zero = basis(2, 0)  # |0>
one = basis(2, 1)   # |1>

# Create a superposition state |psi> = alpha * |0> +  beta * |1>
alpha = 1 / np.sqrt(2)  # Amplitude for |0>
beta = 1 / np.sqrt(2)   # Amplitude for |1>

psi = alpha * zero + beta * one

b.add_states(psi)
b.show()
```

$$ |\psi\rangle = \alpha|0\rangle + \beta|1\rangle $$

```python
from qutip.qip.operations import hadamard_transform

# Apply Hadamard to |0>
psi_hadamard_0 = hadamard_transform() * zero
print("\n|0> after Hadamard:\n", psi_hadamard_0)

# Apply Hadamard to |1>
psi_hadamard_1 = hadamard_transform() * one
print("\n|1> after Hadamard:\n", psi_hadamard_1)
```

```python

b = Bloch()
b.add_states(zero)
b.add_states(one)
b.add_states(psi)
b.add_states(psi_hadamard_1)
b.make_sphere()
b.show()
```

```python
zero = basis(2, 0)  # |0>
one = basis(2, 1)   # |1>

psi_hadamard = hadamard_transform() * zero

b = Bloch()
b.add_states(zero)
b.add_states(psi_hadamard)

b.show()
```
